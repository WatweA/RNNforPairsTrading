```python
import os
import pandas as pd
```

```python
results_data_path = r'data/info/results_regression.txt'

models = [
    'Linear Regression',
    'Support Vector Machine',
    'Random Forest',
    'Adaptive Boost',
    'Gradient Boost',
    'Neural Net',
    'Ensemble of all models'
]

mse_sums = [[], [], [], [], [], [], []]

assert len(mse_sums) == len(models)
```

```python
delimiter = '\n--------------------------------------------------'\
            '\n==================================================\n\n'
x = open(results_data_path, 'r')
split_text = x.read().split(delimiter)
for t in split_text:
    
    
    print(t.split('\n')[0])
    
    all_mse_pos = [i for i in range(len(t)) if t.startswith('MSE', i)]
#     print(all_mse_pos)
    
    for i, mse_pos in enumerate(all_mse_pos):
        
        mse = t[mse_pos+5:mse_pos+13]
        print(i, mse)
        
        mse_sums[i] += [float(mse)]
        
        
#     break
# print(split_text)
```

```python
df = pd.DataFrame(mse_sums).transpose() #, columns = models)
df.columns = models
df
```

```python

boxplot = df.boxplot(column=models, rot=60)
boxplot.set_ylabel('MSE')
```

```python
boxplot = df.drop([9]).boxplot(column=models, rot=60)
boxplot.set_ylabel('MSE')
```

```python
import matplotlib.pyplot as plt
df1 = df.drop(['Linear Regression', 'Support Vector Machine'], axis=1)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = df1.mean().axes[0]
students = df1.mean()
ax.bar(langs,students)
ax.set_xticklabels(df.columns, rotation=40, ha='right')
ax.set_ylabel('Average MSE')
ax.set_title('Average MSE across models')
plt.show()

# df.mean()
```
