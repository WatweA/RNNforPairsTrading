```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
ticker1 = 'ADX'
ticker2 = 'CSX'
```

```python
df1 = pd.read_pickle("data/raw/" + ticker1 + ".zip")  # needs pickle5 compression (python 3.8)
df2 = pd.read_pickle("data/raw/" + ticker2 + ".zip")  # needs pickle5 compression (python 3.8)
```

```python
df1
```

```python
X1 = df1.index
y1 = df1['Adj Close']

X2 = df2.index
y2 = df2['Adj Close']



plt.plot(X1,y1)
plt.plot(X2,y2)
plt.title(ticker1 + "/" + ticker2 + " Stock Price")
plt.ylabel('Price (USD)')
plt.legend([ticker1, ticker2])
plt.show()
```
