This notebook contains a simple script to generate plots to compare two ticker pairs selected by the modified
Gale-Shapley algorithm used in this project to AAPL and JPM over time, to show that the correlational analysis was
successful in a visually interpretable way.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ticker_pairs = [
    ("ADX", "TY"),
    ("ADX", "AAPL"),
    ("ADX", "JPM"),
    ("NNN", "O"),
    ("NNN", "AAPL"),
    ("NNN", "JPM"),
]

for ticker1, ticker2 in ticker_pairs:
    df1 = pd.read_pickle("data/raw/" + ticker1 + ".zip")  # needs pickle5 compression (python 3.8)
    df2 = pd.read_pickle("data/raw/" + ticker2 + ".zip")  # needs pickle5 compression (python 3.8)

    X1 = df1.index
    y1 = df1['Adj Close']
    # y1 = df1['Log Return']

    X2 = df2.index
    y2 = df2['Adj Close']
    # y2 = df2['Log Return']

    plt.plot(X1,y1)
    plt.plot(X2,y2)
    plt.title(ticker1 + " vs. " + ticker2 + " Stock Price")
    plt.ylabel('Price (USD)')
    plt.legend([ticker1, ticker2])
    plt.savefig(f"out/img/{ticker1}-{ticker2}.jpeg")
    plt.show()
```
